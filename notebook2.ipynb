```
from colabtools import nvd3
import numpy
nvd3.CreateScatterPlot(
  [
    {'key': 'Good',
     'values': [nvd3.xy((x, y), label='{:.2f}, {:.2f}'.format(x, y))
      for x, y in 3 * numpy.random.randn(200, 2)]},
    {'key': 'Bad',
     'values': [nvd3.xy((x, y), label='{:.2f}, {:.2f}'.format(x, y))
      for x, y in numpy.random.randn(200, 2)],
    }
  ],
  tooltip_content=nvd3.ValueLabelTooltip(),
  width = 600)
```
